```python
from imageai.Prediction import ImagePrediction
```

```python
# this jupyternotebook uses my local python interpreter instead of anaconda
import sys
print(sys.executable)
```

```python
# follow the tutorials of imageAI to run image prediction
import os
execution_path = os.getcwd()

prediction = ImagePrediction()
prediction.setModelTypeAsResNet()
prediction.setModelPath(os.path.join(execution_path, "resnet50_weights_tf_dim_ordering_tf_kernels.h5"))
prediction.loadModel()
```

```python
predictions, probabilities = prediction.predictImage(os.path.join(execution_path, "images/house.jpg"), result_count=5 )
for eachPrediction, eachProbability in zip(predictions, probabilities):
    print(eachPrediction , " : " , eachProbability)
```
